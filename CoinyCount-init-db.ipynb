Load Coin Database

```python
examples = [pd.read_excel(filename, dtype={"Praegejahr":str}) for filename in glob.glob("temp/muenzen-csv/*.xlsx")]
coin_db = pd.concat(examples, axis=0)
coin_db = coin_db[coin_db['realid'].notnull()]
```

```python
coin_db = coin_db[coin_db['realid'].str.contains('0$')]
coin_db['realid'] = coin_db['realid'].str.replace(r'0$', '')
```

```python
coin_db_renamed = coin_db.rename(columns = {
    'realid' : 'Artikelnr'
}).set_index('Artikelnr')
coin_db_renamed = coin_db_renamed[['Land', 'Praegejahr', 'name', 'CMD', 'groupid', 'Nominale', 'weight', 'Auflage', 'company', 'description', 'description_en', 'specification']]
```

```python
coin_db_renamed.to_excel('Münzdatenbank.xlsx', sheet_name='Datenbank')
```
